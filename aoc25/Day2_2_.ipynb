Possible cases:

the particle (repeated part) is even:
    repeated even number of times -->even number
    repeated odd number of times --> even number
    
the particle is odd:
    repeated odd number of times --> odd number
    repeated even number of times --> even number

If even: 
    check if halving gives pattern

    if not --> hardcoded: check if divided by 3/5... gives pattern (max length of ids is 9)

    if not --> valid id

If odd: 

    check if all digits are the same
    check if divided by 1/2/3/5 gives pattern

    



```python
def is_invalid(s):
    if len(s)%2==1: #if it is an odd number
        return False

    middle=int(len(s)/2) #calculating the middle point

    first_half = s[0:middle] #from first index until middle (last index is excluded in Python)
    second_half =s[middle:] #from middle until end (indexing starts with 0)
    if first_half == second_half:
        return True
    else:
        return False
```
