```python
def is_invalid(s):
    if len(s)%2==1: #if it is an odd number
        return False

    middle=int(len(s)/2) #calculating the middle point

    first_half = s[0:middle] #from first index until middle (last index is excluded in Python)
    second_half =s[middle:] #from middle until end (indexing starts with 0)
    if first_half == second_half:
        return True
    else:
        return False
    
```
