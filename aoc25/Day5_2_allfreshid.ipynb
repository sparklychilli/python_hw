1. read in input, keep only range lines (if line contains "-", add to range list, else continue)
2. loop through ranges, add ids to fresh ids (only add if not in list already) (alternatively, make set at the end, but that might be too much data for python?)

```python
def read_file_into_lines(file_path):
    """
    A simple function to read a file into
    a list of lines, removing the newline
    character at the end.

    Arguments:
    ==========
    file_path: str
        The location of a text file to read.

    Returns:
    ========
    list[str]
        A list that contains all stripped lines
        of the input text file.
    """
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
    
```

```python
def get_fresh_ids(puzzle_input):
    '''reads lines with ingredient ranges, discards other lines; computes a list of unique fresh ingredient ids'''

    fresh_ids = []
    fresh_ranges = []

    for line in puzzle_input:
        if "-" in line: #if it is formated as a range like 13-14
            fresh_ranges.append(line)
        else:
            continue
    
    for f in fresh_ranges:
        start, stop = f.split("-")
        for ing in range(int(start), int(stop)+1):
            if ing in fresh_ids:
                continue
            else:
                fresh_ids.append(ing)
                
    fresh_ids.sort()

    return fresh_ids
            
```
