```python
def read_file_into_lines(file_path):
    """
    A simple function to read a file into
    a list of lines, removing the newline
    character at the end.

    Arguments:
    ==========
    file_path: str
        The location of a text file to read.

    Returns:
    ========
    list[str]
        A list that contains all stripped lines
        of the input text file.
    """
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
    
```

```python
#simplifying ranges:

def overlapping_ranges(range1, range2):
    #check if two ranges overlap or are contained in each other
    start1 = range1[0]
    stop1 = range1[1]
    start2 = range2[0]
    stop2 = range2[1]

    if start1 <= start2 and stop1 >= stop2:
        return True
    if start1 >= start2 and stop1 <= stop2:
        return True
    if start1 <= start2 and stop1 >= start2:
        return True
    if start2 <= start1 and stop2 >= start1:
        return True
    else:
        return False
```

```python
def merge_ranges(range1, range2):
    #if two ranges overlap, merge the two by setting the start to the smaller start and the stop to the bigger stop
    if overlapping_ranges(range1, range2):
        start1 = range1[0]
        stop1 = range1[1]
        start2 = range2[0]
        stop2 = range2[1]

        merged = [min(start1, start2), max(stop1, stop2)]
        return merged
```

```python
def parse_ranges(filepath):
    #takes only the range lines from the input file and separates each ranges into a list with start and stop, than adds them to a list of lists
    puzzle_input = read_file_into_lines(filepath)
    
    range_lines = []
    fresh_ranges = []
    

    for line in puzzle_input:
        if "-" in line: #if it is formated as a range like 13-14
            range_lines.append(line)
        else:
            continue
    
    for line in range_lines:
        start, stop = line.split("-")
        fresh_ranges.append((start, stop))
        
    return fresh_ranges
```
