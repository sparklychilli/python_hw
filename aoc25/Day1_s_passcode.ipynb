Dial with 100 locations, 0-99

1 number, 1 click

can turn left & right
left--> lower number
right --> higher number

turns continously (e.g. from 99 to 0) in both directions

start = 50

For example, suppose the attached document contained the following rotations:

L68
L30
R48
L5
R60
L55
L1
L99
R14
L82
Following these rotations would cause the dial to move as follows:

The dial starts by pointing at 50.
The dial is rotated L68 to point at 82.
The dial is rotated L30 to point at 52.
The dial is rotated R48 to point at 0.
The dial is rotated L5 to point at 95.
The dial is rotated R60 to point at 55.
The dial is rotated L55 to point at 0.
The dial is rotated L1 to point at 99.
The dial is rotated L99 to point at 0.
The dial is rotated R14 to point at 14.
The dial is rotated L82 to point at 32.

```python
sample_instuctions = (
"L68",
"L30",
"R48",
"L5",
"R60",
"L55",
"L1",
"L99",
"R14",
"L82")
```

```python
def turn_password_dial(dial_position, instruction_list):
    #return the passcode nr for a password dial when given the starting position and a list of instructions formatetted "Lnn", "Ln" or "Rnn", "Rn"
    for turn in instruction_list:
        direction = turn[0]
        steps = int(turn[1:])
                    
        if direction == "L":
            dial_position = dial_position - steps
            if dial_position > 99:
                dial_position = dial_position - 100
            if dial_position < 0:
                dial_position = dial_position + 100
                
        if direction == "R":
            dial_position = dial_position + steps
            if dial_position > 99:
                dial_position = dial_position - 100
            if dial_position < 0:
                dial_position = dial_position + 100
    return(dial_position)         
```

```python
turn_password_dial(50, sample_instuctions)
```
