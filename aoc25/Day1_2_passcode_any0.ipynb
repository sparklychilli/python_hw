Like in the first part, we dial to get the password. But now the password is the number of times the dial hits 0 while turning (not only final dial positions after a turn)

```python
sample_instructions = (
"L68",
"L30",
"R48",
"L5",
"R60",
"L55",
"L1",
"L99",
"R14",
"L82")
```

```python
#Reading in the instruction textfile

def read_file_into_lines(file_path):
    """
    A simple function to read a file into
    a list of lines, removing the newline
    character at the end.

    Arguments:
    ==========
    file_path: str
        The location of a text file to read.

    Returns:
    ========
    list[str]
        A list that contains all stripped lines
        of the input text file.
    """
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
```

```python
# getting the password instructions as a list

password_instructions = read_file_into_lines("./aoc_day1.txt")
```

```python
def dial_password_all_hits(dial_position, instruction_list):
    '''return the passcode (times zero is hit) for a password dial when given the starting position
    and a list of instructions formatetted "L" or "R" followed by one to several numbers'''
    password=0
    for turn in instruction_list:
        #print(turn)
        direction = turn[0]
        steps = int(turn[1:])
        if steps > 99: #correct for instructions that make a full turn or more
            #print(steps)
            password+= steps//100 #account for all times 0 is hit with full turns
            #print(steps//100, "hits")
            steps = steps -(steps//100)*100 #calculate the number of full turns and substract it
            

        if direction == "L":
            if dial_position == 0: #when starting from 0, don't count going over 0 as hit
                dial_position = dial_position - steps
                if dial_position > 99:
                    dial_position = dial_position - 100
                if dial_position < 0:
                    dial_position = dial_position + 100
                continue
            if dial_position > 0:
                dial_position = dial_position - steps
                if dial_position > 99:
                    dial_position = dial_position - 100
                    password += 1 #account for going over 0
                    #print("hit")
                if dial_position < 0:
                    dial_position = dial_position + 100
                    password+=1 #account for going over 0
                    #print("hit")
        if direction == "R":
            if dial_position == 0:
                dial_position = dial_position + steps
                if dial_position > 99:
                    dial_position = dial_position - 100
                if dial_position < 0:
                    dial_position = dial_position + 100
                continue
            if dial_position >0:
                dial_position = dial_position + steps
                if dial_position == 100:
                    dial_position = 0
                if dial_position > 99:
                    dial_position = dial_position - 100
                    password += 1 #account for going over 0
                    #print("hit")
                if dial_position < 0:
                    dial_position = dial_position + 100
                    password += 1 #account for going over 0
                    #print("hit")

        if dial_position  == 0:
            password += 1
            #print("hit")

    return(password)       
    

```
