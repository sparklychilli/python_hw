For each Line(bank):
    -->find the biggest possible joltage; 
    first number=first position, second number=second position;
    first and second position have to be consecutive in the bank but not exactly next to each other
            -->thus, the first position has to be the next to last battery latest
            --->the second position can only be after the first position
Calculate the sum of all the max. output voltages

```python
def get_max_joltage(bank):
    #loop through a bank of batteries and find the max voltage number
    first_position = 0
    second_position = 0              
    first_index = 0
    joltage = ""
    
    #enumerate to loop through index and values
    for ind, jr1 in enumerate(bank[:-1]):
        if int(jr1) > first_position:
            first_position = int(jr1)
            first_index = ind
        elif int(jr1) <= first_position: #smaller or equal to sign?
            continue   
        joltage = str(first_position) #needs to be a string to concatenate later!
    
        #print(ind, first_position, joltage)
    
    for jr2 in bank[(first_index + 1):]: #loop through rest of bank
        if int(jr2) > second_position:
            second_position = int(jr2)
            # print(second_position)
        elif int(jr2) <= second_position: 
            continue
    joltage+= str(second_position)
    
    joltage=int(joltage)
    return joltage
```

```python
example_banks = (
"987654321111111",
"811111111111119",
"234234234234278",
"818181911112111")
```

```python
total_joltage=0
for bank in example_banks:
    total_joltage+=(get_max_joltage(bank))
print(total_joltage)
```

```python
def get_total_joltage(banks):
#calculate the total joltage of a list of banks
    total_joltage=0
    for bank in banks:
        total_joltage+=(get_max_joltage(bank))
    print(total_joltage)
```

```python
#read file into lines function

def read_file_into_lines(file_path):
    """
    A simple function to read a file into
    a list of lines, removing the newline
    character at the end.

    Arguments:
    ==========
    file_path: str
        The location of a text file to read.

    Returns:
    ========
    list[str]
        A list that contains all stripped lines
        of the input text file.
    """
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
```
