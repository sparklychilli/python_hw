Problem
In DNA strings, symbols 'A' and 'T' are complements of each other, as are 'C' and 'G'.

The reverse complement of a DNA string s
 is the string sc
 formed by reversing the symbols of s
, then taking the complement of each symbol (e.g., the reverse complement of "GTCA" is "TGAC").

Given: A DNA string s
 of length at most 1000 bp.
 -

Return: The reverse complement sc
 of s
 -
.

Sample Dataset
AAAACCCGGT
Sample Output
ACCGGGTTTT


There are several steps to this:
1. replace each base with the complementing base
    * e.g. through several steps with the replace function)
    * but how do I make sure not to double replace?
    * If I take the whole string and replace all A with T and then all T with A, I end up with only A
      --> this means I have to work indivividual base by base, e.g working with a range and a for loop and then concatenate
2. reverse the string
    * check if there is a function for that?
    * Alternatively, I could try to concatenate in reverse order, I think that should work? so instead of revc = revc + new_base I can do revc = new_base + revc
      --> let's try that

```python
# 1. replace each base with complementing one
DNA = "AAAACCCGGT" #asigning DNA string to a variable
REVC = "" #making an empty string that will hold my REVC bases

for r in range(0, len(DNA),1): #loop through all positions of the string
    if DNA[r] == "A": #making an if/elif loop to assign all possible complements that the base could have
        new_base = "T"
    elif DNA[r] == "T":
        new_base = "A"
    elif DNA[r] == "G":
        new_base = "C"
    elif DNA[r] == "C":
        new_base = "G"
    #print(new_base)
    
# 2. make a reverse string

    REVC = new_base + REVC #concatenating the new_base to the REVC string
    
print(REVC) #print the string outside the loop to have the endresult
```

```python
#Check if my result and the sample output match up
sample_output = "ACCGGGTTTT"

if REVC == sample_output:
    print ("Succesfully made reverse compliment string! Good job <3")
```
