Given: Three positive integers k
, m
, and n
, representing a population containing k+m+n
 organisms: 
 k...homozygous dominant 
 m...heterozygous
 n...homozygous recessive.

Return: The probability that two randomly selected mating organisms will produce an individual possessing a dominant allele (and thus displaying the dominant phenotype). Assume that any two organisms can mate.

probability of desired outcomes: 
first parent is homozygous dominant: --> offspring phenotype is dominant no matter the second parent: k/N

(m/N)*(k/(N-1))*


My probability of dominant Phenotype = 1 - recessive Phenotype

p(recessive Phenotype) = ((m**2) - m + (4*m*n) - (4*n) + (4*(n**2)))/((N**2) - N)
p(dominant Phenotype) = 1 - p(recessive)

Sample Dataset
2 2 2
Sample Output
0.78333
