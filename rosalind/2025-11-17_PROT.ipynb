#Translating RNA into Protein#

##Problem##

The 20 commonly occurring amino acids are abbreviated by using 20 letters from the English alphabet (all letters except for B, J, O, U, X, and Z). Protein strings are constructed from these 20 symbols. Henceforth, the term genetic string will incorporate protein strings along with DNA strings and RNA strings.

The RNA codon table dictates the details regarding the encoding of specific codons into the amino acid alphabet.

###Given: An RNA string s
 corresponding to a strand of mRNA (of length at most 10 kbp).

###Return: The protein string encoded by s

###Sample Dataset
AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA
###Sample Output
MAMAPRTEINSTRING

1. find all the codons in s (split in groups of three & 2. for every codon find corresponding AA

I want to slice the string every three positions...
How do I slice of the first three?
How do I slice of n^th position?

how often do I have to slice?

s.len()/3
or: (last index number + 1)/3


```python
s = "AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA"
```

```python
#printing the first codon: 
cod_1 = s[0:3]
print(cod_1)
```

```python
#print the fourth codon:
#start of 4th codon: 0 + 3**3, so (codon nr -1)**3
#start of 5th codon: 0 + 4**3
#my index for the start of the nth codon is (n-1)**3 --> this would not work for the first position though
#my index for the end would be (n-1)**3 + 3
cod_4 = s[9:12]
print(cod_4)
```

```python
#Getting the codons and assigning amino acids
last_index = len(s) - 1
for i in range (0, last_index, 3):#Making a for loop to print all codons:
    codon = s [i:(i+3)]
    print(codon)
    aminoacid = codon_RNA_dict[codon]
    print (aminoacid)
```

```python
#Now I want to print the whole protein string together, so concatenate each aminoacid to the previous one
last_index = len(s) - 1
protein_string = "" #making an empty string that I will add my aminoacids to
for i in range (0, last_index, 3):#Making a for loop to print all codons:
    codon = s [i:(i+3)] #getting the codon
    aminoacid = codon_RNA_dict[codon] #getting the corresponding amino acid
    protein_string = protein_string + aminoacid # concatenating each new codon to my protein string

print(protein_string) #printing the final protein string (outside the loop so it is done after the last iteration)
```
