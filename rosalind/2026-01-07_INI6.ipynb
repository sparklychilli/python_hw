# Given: 
A string s of length at most 10000 letters.

# Return: 
The number of occurrences of each word in s, where words are separated by spaces. Words are case-sensitive, and the lines in the output can be in any order.

```python
#sample_imput:
"We tried list and we tried dicts also we tried Zen"
```

```python
string="We tried list and we tried dicts also we tried Zen"

words=list(set(string.split(" ")))
wordcount={}

for word in words:
    count = string.count(word)
    wordcount[word]=count
    #print(word, count)#this would be the straightforward way to print in loop without dictionary)
    
for key, value in wordcount.items():
    print(key, value)
```
