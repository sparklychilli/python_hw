An RNA string is a string formed from the alphabet containing 'A', 'C', 'G', and 'U'.

Given a DNA string t
 corresponding to a coding strand, its transcribed RNA string u
 is formed by replacing all occurrences of 'T' in t
 with 'U' in u
.

Given: A DNA string t having length at most 1000 nt.
-

Return: The transcribed RNA string of t
-

The description pretty much tells me what to do:
1. in a DNA string, find all ocurrences of T
2. replace those occurences with U
   
-->sounds like a job for the replace method
   


Sample Dataset
GATGGAACTTGACTACGTAAATT
Sample Output
GAUGGAACUUGACUACGUAAAUU

```python
s = "GATGGAACTTGACTACGTAAATT" #saving the sample DNA string
```

```python
s.replace("T", "U") #replacing T with U, thus creating an RNA string
```

```python
t = s.replace ("T", "U") #saving my result as a new string
```

```python
'''Let's see if my result and the sample output match up'''
sample_output = 'GAUGGAACUUGACUACGUAAAUU'

if t == sample_output:
    print("success")
```
