Given: Two positive integers a
 and b
 (a<b<10000
).

Return: The sum of all odd integers from a
 through b
, inclusively.


```python
'''
Example:
Imput:
100 200

Output:
7500
'''
```

```python
a=100
b=200

output=0

for i in range(a,b+1):
    if i==1:
        output+=i
        continue
    if i%2==0:
        continue
    if i%2==1:
        output+=i
output
    
```
